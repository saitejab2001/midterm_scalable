# Using duckdb to run SQL commands

```python
import duckdb as dd
```

**Selecting all the column from the first databse "df1"**

```python
query_1 = "SELECT * FROM df1"
result = dd.query(query_1).to_df()
print(result)
```

**Total number of incident for specific boro**

```python
query_2 = "SELECT BORO, COUNT(INCIDENT_KEY)  FROM df1 GROUP BY BORO"
result = dd.query(query_2).to_df()
print(result)
```

Races commited crimes

```python
query_3 = "SELECT VIC_RACE, COUNT(INCIDENT_KEY) FROM df1 GROUP BY VIC_RACE"
result = dd.query(query_3).to_df()
print(result)
```

**Joining two tables using outer join to consider all the entries from database df1 and df2**

```python
query_4 = "SELECT * FROM df1 FULL OUTER JOIN df2 ON df1.INCIDENT_KEY=df2.INCIDENT_KEY;"
print(dd.query(query_4).to_df())
```

**But as we have to stack the data set because of the common attributes we will use union**

```python
query_5 = "SELECT * FROM df1 UNION ALL SELECT * FROM df2;"
df = dd.query(query_5).to_df()
print(df)
```

**Converting file into csv so that it will get xconverted into pandas dataframe format**

```python
df.to_csv("/content/Joined_database")
df = pd.read_csv("/content/Joined_database")
df.head()
```

**Cleaning our finalized dataset**

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

**Dropping the na values and some inaccurate values**

```python
#Dropping na values
df.dropna(axis=0, inplace=True)

#Replacing the U = Uknown values with mode of 'perp_sex'
df['PERP_SEX'].replace(to_replace='U',value=df['PERP_SEX'].mode()[0], inplace=True)

#Now replacing inaccurate age groups with the mode of it
inaccurate_age = ['1020','940','224','1028']

for val in inaccurate_age:
    df['PERP_AGE_GROUP'].replace(to_replace=val,value=df['PERP_AGE_GROUP'].mode()[0], inplace=True)

#Now dropping duplicate values
df.drop_duplicates(inplace=True)
```

```python
#Converting following columns dtype into Datetime
from datetime import date
df['OCCUR_DATE'] = pd.to_datetime(df['OCCUR_DATE'])
df['OCCUR_TIME'] = pd.to_datetime(df['OCCUR_TIME'])

# This will create a column in the data base called as "month", And will extract only the months from "OCCUR_DATE" column
df['Month'] = df['OCCUR_DATE'].dt.month

# Count the number of arrests for each month
incident_month = df['Month'].value_counts().sort_index()
print(incident_month)
```

```python
# Here we are using .plt method to plot the graph first mentioning index of the data set then the actual "incident_month dataset"
plt.plot(incident_month.index, incident_month, marker='.', color='blue', linestyle='-', linewidth=1)
plt.title('INCIDENT PER MONTH')
# Then we are assigning numbers to belowe specified string i.e 1='jan', 2='feb'.
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
# Here we are labeling x and y axis
plt.xlabel('MONTH')
plt.ylabel('NUMBER OF INCIDENTS')
# Again changing the figure size
plt.figure(figsize=(12, 6))
# Then finally plotting it
# This data gives the number incident happened per month
plt.show()
```

**Lets plot the location of crime which occured in manhattan**

```python
# We have imported a new library used for leaflet map plotting
import folium

#Now Filtering the dataset for crime which occured in manhattan
map_df = df[df['BORO'] == 'STATEN ISLAND']

# finding mean to create a base map
inc_map = folium.Map(location=[map_df['Latitude'].mean(), map_df['Longitude'].mean()], zoom_start=12)

# Adding markers for each shooting spot
for a,r in map_df.iterrows():
    folium.Marker(
        location=[r['Latitude'], r['Longitude']],
        popup=f"{r['BORO']}\n{r['PERP_SEX']}",
        icon=folium.Icon(color='red', icon='gun', prefix='fa')
    ).add_to(inc_map)

# saving file into .HTML
inc_map.save('shooting_incident_map_sample.html_')

# Will display map
inc_map
```

```python
#Here w are importing plotly library and using the module express for geographical plotting
import plotly.express as px

#Creating a new column which gives us hour of the day
df['HOUR']=df['OCCUR_TIME'].dt.hour

#Here we are plotting on the map on the based of thegiven data
fig = px.scatter_mapbox(df, lat='Latitude', lon='Longitude', color='BORO',size = 'HOUR',
                        color_continuous_scale=px.colors.cyclical.IceFire, size_max = 15,
                        zoom=10,mapbox_style='carto-positron')
fig.show()
```

```python
data = pd.read_csv("/content/park crime NY.csv")
data.head()
```

```python
data.info()
```

```python
data.dropna(axis=0, inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(data.isnull())
```

```python
borough_crimes = data.groupby('BOROUGH')[['MURDER', 'RAPE', 'ROBBERY', 'FELONY ASSAULT',
                                          'BURGLARY', 'GRAND LARCENY', 'GRAND LARCENY OF MOTOR VEHICLE']].sum()

borough_crimes.plot(kind='barh',stacked=True,figsize=(12, 6), colormap='viridis')
plt.title('Sum of Major Crimes by Borough')
plt.xlabel('Borough')
plt.ylabel('Crime Count')
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
plt.xticks(rotation=45, ha='right', fontsize=10)
sns.barplot(data,x='CATEGORY', y='TOTAL', hue='BOROUGH', estimator='sum', palette='rocket')
plt.title('Total number of crimes commited by category')
plt.xlabel('Category')
plt.ylabel('Crime Count')
plt.show()
```

```python
# We are using groupby function to group the data according to the borough and then calculating sum of each instance
db_category = data.groupby('CATEGORY').sum()
# Printing the head of the databse
db_category.head()
# Setting the size of the figure aloing side x and y axis
sns.set_theme(rc={'figure.figsize':(12,8)})
# Here assigning the variable with all the relevant data for plotting purpose, where total crime is shown across the type & size of the park.
bar = sns.barplot(x='TOTAL', y='CATEGORY', data=db_category, palette='rocket_r')
# Adding labels and plotting it
bar.set(xlabel='AREAS',ylabel='COUNTS')
```

```python
# Here we are using violin graph to represents a classified type of crime and number of occurence of it
data1 = data.drop(['TOTAL', 'SIZE (ACRES)'], axis=1)
sns.set_theme(rc={'figure.figsize':(25,6)})
sns.violinplot(data=data1)
```
